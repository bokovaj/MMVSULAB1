```python
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
url='https://raw.githubusercontent.com/bokovaj/MMVSULAB1/main/spb.csv'
df=pd.read_csv(url)
print(df)
years=np.array(df['Years'])
population=np.array(df['Population'])
plt.plot(years,population)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of spb since 1949')
plt.show()
     
```
