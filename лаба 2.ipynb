```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def dydt(x,t):
  k = 0.0044
  return k*x
t = [1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005]
solve = odeint(dydt, 5007469, t=t)
solve = solve.flatten()
%matplotlib inline 
v = [5007469, 4986405, 4942891, 4881563, 4845407, 4820213, 4806641, 4783982, 4770897, 4741923, 4714844, 4688414, 4656474, 4662038, 4686491   ]
plt.plot(t, solve, label = r"x′=kx")
plt.plot(t, v, label = r"Население г.Санкт-Петербург ") 
plt.grid(True)
plt.legend()
plt.show()
```
