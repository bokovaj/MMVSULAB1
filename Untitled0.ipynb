```python
import matplotlib.pyplot as plt 
import pandas as pd
from scipy.integrate import odeint 


data = {'Year': [1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000,2001,2002,2003],
        'Population': [5007469, 4986405, 4942891, 4881563, 4845407, 4820213, 4806641, 4783982, 4770897, 4741923, 4714844, 4688414, 4656474 ]}
df = pd.DataFrame(data)

plt.plot(df['Year'], df['Population'])

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of Saint Petersburg since 1991')

plt.show()

```

```python

import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def dydt(x,t):
  k = -0.13
  return k*x
t = np.linspace(0,10,100)
solve = odeint(dydt, 1, t=t)
solve = solve.flatten()
%matplotlib inline 
plt.plot(t, solve, label = r"x' = kx")

plt.grid(True)
plt.legend()
plt.show()
```
